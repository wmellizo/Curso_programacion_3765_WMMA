```python
# Ejercicio de Refuerzo 1
sueldo = float(input("Digite el valor del sueldo mensual: "))
venta1 = float(input("Digite el valor de la Venta No. 1: "))
venta2 = float(input("Digite el valor de la Venta No. 2: "))
venta3 = float(input("Digite el valor de la Venta No. 3: "))
com1 = round(venta1 * 0.1, 2)
com2 = round(venta2 * 0.1, 2)
com3 = round(venta3 * 0.1, 2)
comision = round((com1 + com2 + com3), 2)
Totalsueldo = sueldo + comision
print(f"\nEl valor de cada una de las comisiones por ventas es: \nComisión 1 $ {com1:,.0f} \nComisión 2 $ {com2:,.0f} \nComisión 3 $ {com3:,.0f} \n ")
print(f"El valor total de las Comisiones por ventas son: $ {comision:,.0f}")
print(f"\nEl sueldo total devengado por el vendedor es: Sueldo Basico $ {sueldo:,.0f} + Comisiones $ {comision:,.0f} = $ {Totalsueldo:,.0f}")
```

```python
#Ejercicio de Refuerzo 2
venta = float(input("Digite el Valor Neto de la venta: "))
desc = round(venta * 0.15, 2)
total = venta - desc
print(f"El valor Neto de la venta es: $ {venta:,.0f}")
print(f"El valor del descuento en la venta es: $ {desc:,.0f}")
print(f"El valor Total de la venta es: \n$ {venta:,.0f} Neto Venta \n- $ {desc:,.0f} Descuento \n$ {total:,.0f} Valor Total de la Venta")
```

```python
#Ejercicio de Refuerzo 3
nota1 = float(input("Digite la nota del parcial 1: "))
nota2 = float(input("Digite la nota del parcial 2: "))
nota3 = float(input("Digite la nota del parcial 3: "))
promedio = round((nota1+nota2+nota3)/3, 1)
print(f"La nota obtenida por el estudiante es: {promedio}, notas parciales de {nota1}, {nota2} y {nota3}")
```

```python
#Ejercicio de Refuerzo 4
hombres = int(input("Digite la cantidad de hombres del grupo: "))
mujeres = int(input("Digite la cantidad de mujeres del grupo: "))
total = hombres + mujeres
Por_h = (hombres/total)*100
Por_m = (mujeres/total)*100
print(f"El grupo se compone de {total} personas, el {Por_h:,.0f}% de hombres y el {Por_m:,.0f}% de mujeres.")
```
