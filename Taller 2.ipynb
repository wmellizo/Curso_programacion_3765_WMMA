```python
# Taller 2
# Registrar los datos del trabajador por parte del usuario
No_identificacion = int(input("Digite el número de documento de identidad: "))
Nombre = input("Digite el Nombre: ")
Apellidos = input("Digite los Apellidos: ")
Direccion = input("Ingrese la Dirección: ")
Telefono = int(input("Registre el número de celular: "))
Edad = int(input("Cual es su edad: "))
Estado = input("Estado civil: ")
No_hijos = int(input("Número de hijos: "))
Estatura = int(input("Registre la estatura en centímetros: "))
dia_contrato = int(input("Registre el día de inicio del contrato: "))
mes_contrato = input("Registre el mes de inicio de contrato, ejemplo Noviembre: ")
ano_contrato = int(input("Registre el año de inicio de contrato, ejemplo 2001: "))
sueldob = float(input("Cual es el sueldo del trabajador: "))
dias_trabajados = int(input("Cantidad de días laborados entre 0 y 30: "))

# Mostrar en pantalla los datos capturados
print("\nNúmero de documento de identidad: ", No_identificacion)
print("Nombres: " + Nombre)
print("Apellidos: " + Apellidos)
print("Dirección: " + Direccion)
print("Número de Celular: ", Telefono)
print("Edad: ",Edad)
print("Estado Civil: " + Estado)
print("Cantidad de hijos: ", No_hijos)
print("Estura en centímetros: ", Estatura)
print("Fecha de Inicio del Contrato: día", dia_contrato, "Mes " + mes_contrato, "Año", ano_contrato)
print(f"Sueldo Básico: $ {sueldob:,.0f}")
print("Total de días laborados: ", dias_trabajados)

# Inicio Taller 2
# Punto 1
if Edad >14 and Edad <62: 
  print("\nLa Edad es apta para laborar en Colombia.")
  sueldo = sueldob
elif Edad >55:
  bono = Edad*0.05
  sueldo = sueldob + bono
  print(f"\nEl valor de su bono prepensional es: $ {bono:,.0f}")
else:
  print(f"\nLa edad de {Edad} años no es apta para ser trabajador en Colombia.")

# Punto 2
if Estado == "Casado" and No_hijos >0:
  paseo = "SI"
  print(f"\nEl trabajador {paseo} tiene derecho al Paseo anual de Diciembre.")
else:
  paseo = "NO"
  print(f"\nEl trabajador {paseo} tiene derecho al Paseo anual de Diciembre.")

# Punto 3
if sueldob >= 1000000 and sueldob <= 1500000:
  comision = round(sueldob * 0.02, 2)
  sueldo = sueldo + comision
  print(f"'nEl valor de la Comisión por salario es $ {comision:,.2f}")
if sueldob > 1500000 and sueldob <= 2000000:
  comision = round(sueldob * 0.05, 2)
  sueldo = sueldo + comision
  print(f"\nEl valor de la Comisión por salario es $ {comision:,.2f}")
if sueldob < 1000000 or sueldob > 2000000:
  print("\nEl trabajador no está comisionando en salario")

# Punto 4
if dias_trabajados > 20 and sueldob < 1000000:
  print("\nEl Trabajador tiene Derecho a Bono de Alimentación.")
else:
  print("\nEl Trabajador NO tiene Derecho a Bono de Alimentación.")

print(f"\nEl total devengado por el trabajador es: $ {sueldo:,.0f} y su sueldo básico es $ {sueldob:,.0f}")
```
