```python
#Ejercicio 1
print("Los números pares del 0 al 100 son:")
for a in range(0,101,2):
  print(a)
```

```python
#Ejercicio 2
print("Hoy haremos las tablas de multiplicar del 0 hasta el 20.")
n = int(input("¿de qué número quieres ver el cálculo?: "))
for a in range(21):
  if a<10:
    if (n*a)<10:
      print(n, " X  ", a, " =  ", n*a)
    else:
      print(n, " X  ", a, " = ", n*a)
  else:
    print(n, " X ", a, " = ", n*a)
```

```python
#Ejercicio 3
edad = int(input("Digite su edad: "))
if edad > 0: 
  for a in range(1,edad+1):
    if a < 10:
      print(f"Usted ha cumplido los siguientes años:  {a}.")
    else:
      print(f"Usted ha cumplido los siguientes años: {a}.")
else:
  print("La Edad ingresada no es correcta")
```

```python
#Ejercicio 4
print("Mostraremos los números impares desde 1 hasta el número que se indique.")
n = int(input("Digite el número: "))
for a in range(1,n+1,2):
  if a == 1:
    print(f"El primer número impar es: {a}.")
  if a < 10 and a > 1:
    print(f"El siguiente número impar es:  {a}.")
  if a > 10:
    print(f"El siguiente número impar es: {a}.")
```

```python
#Ejercicio 5
print("Las sumas consecutivas de los números pares del 0 al 100 son:")
acum = 0
b = 1
for a in range(0,101,2):
  acum = acum + a
  if b < 10:  
    print(f"Suma  {b}. ",acum, " + ", a, " = ", a + acum)
    b = b + 1
  else:
    print(f"Suma {b}. ",acum, " + ", a, " = ", a + acum)
```
